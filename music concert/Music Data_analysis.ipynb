## 1. list the customers that are from brazil and USA

```python
rd(''' select customer_id, first_name || ' ' || last_name as name,
country 
from customer
where country in ('Brazil', 'USA')
group by 3, 2
order by 3 desc ''')
```

## 2. list the Employees who are Sales Agents

```python
rd('''select employee_id, first_name || ' ' || last_name as full_name , title
from employee
where employee.title like '%sales%agent%' ''')
```

## 3. list the invoices associated with each sales agent. The resultant table should include the Sales Agent's full name

```python
rd('''select e.employee_id, e.first_name || ' ' || e.last_name as full_name , e.title, i.invoice_id
from customer c
join invoice i on c.customer_id = i.customer_id
join employee e on c.support_rep_id = e.employee_id
where e.title like '%sales%agent%'
''')
```

## 4. Which country has most number of invoices?

```python
rd('''select i.billing_country, count(i.billing_country) as num_invoices
from invoice i
group by 1
order by 2 desc''')
```

## 5. Which city has the best customers? 

```python
rd('''select billing_city, sum(total) as total_bill_city
from invoice
group by 1
order by 2 desc
limit 1''')
```

## 6. what is people's favourite Genre?

```python
rd('''select g.name as genre, count(c.customer_id) as people_choice 
from customer c 
join invoice i on c.customer_id = i.customer_id
join invoice_line il on i.invoice_id = il.invoice_id
join track t on t.track_id = il.track_id
join genre g on g.genre_id = t.genre_id
group by 1
order by 2 desc''')
```

## 6. Get email ids of the people who like rock music
email id required for inviting people to rock music concert

```python
rd('''select distinct c.email, c.first_name, c.last_name
from customer c
join invoice i on c.customer_id = i.customer_id
join invoice_line il on il.invoice_id = i.invoice_id
where il.track_id in (select track_id 
from track t
join genre g on t.genre_id = g.genre_id 
where g.name like '%rock%')
order by 1''')
```

## 7. Who is writing the rock music?
Now that we know that our customers love rock music, we can decide which musicians to invite to play at the concert.Let's invite the artists who have written the most rock music in our dataset. Write a query that returns the Artist name and total track count of the top 10 rock bands

```python
rd('''select a.name as Artist, count(t.track_id) as total_tracks 
from artist a 
join album ab on ab.artist_id = a.artist_id
join track t on t.album_id = ab.album_id 
join genre g on t.genre_id = g.genre_id
where g.name like '%rock%' 
group by 1
order by 2 desc
limit 10 ''')
```

## 8. Revenue for the artist
Find the revenue collected for the artist and sort it from high to low. This can help us in understanding the most in demand artist

```python
rd('''select a.name as Artist, sum(il.unit_price * il.quantity) as revenue 
from artist a
join album ab on a.artist_id = ab.artist_id 
join track t on t.album_id = ab.album_id
join invoice_line il on il.track_id = t.track_id
group by 1
order by 2 desc''')
```
